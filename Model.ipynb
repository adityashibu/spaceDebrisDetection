## Using YOLO11

### Training the YOLO model

```python
from ultralytics import YOLO

model = YOLO("yolo11n.pt")
model.train(data=r"C:\Users\Aditya Uni\spaceDebrisDetection\Space-debris-2\data.yaml", epochs=30, batch=16, imgsz=640)
```

### Evaluating the model

```python
results = model.val(data=r"C:\Users\Aditya Uni\spaceDebrisDetection\Space-debris-2\data.yaml")
print(results)
```

### Testing the Model

```python
output_dir = r"C:\Users\Aditya Uni\spaceDebrisDetection\Space-debris-2\results"

# Run inference and save predictions to the folder
results = model.predict(
    source=r"C:\Users\Aditya Uni\spaceDebrisDetection\ample-synthetic-1\train\images",
    imgsz=640,
    save=True,
    project=output_dir,
    name='inference_results',
)

print(f"Predictions saved to: {output_dir}/inference_results")
```
